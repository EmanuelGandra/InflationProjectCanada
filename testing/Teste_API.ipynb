```python
from statsmodels.tsa.seasonal import STL
# garantir ordenação correta
teste_df_cpi_55 = (
    teste_df_cpi_55
    .sort_values(["Products and product groups", "REF_DATE"])
    .copy()
)

teste_df_cpi_55
```

```python
def teste_ajuste_sazonal_stl(serie, periodo=12):
    serie = serie.dropna()
    if len(serie) < 2 * periodo:
        return serie
    stl = STL(
        serie,
        period=periodo,
        robust=True
    )
    res = stl.fit()

    # série ajustada = original - componente sazonal
    serie_sa = serie - res.seasonal

    return serie_sa
```

```python
lista_sa = []

for prod in teste_componentes_55:
    df_prod = teste_df_cpi_55[
        teste_df_cpi_55["Products and product groups"] == prod
    ]

    serie = (
        df_prod
        .set_index("REF_DATE")["VALUE"]
        .astype(float)
        .sort_index()
    )

    serie_sa = teste_ajuste_sazonal_stl(serie)

    df_sa = (
        serie_sa
        .rename("indice_sa")
        .reset_index()
    )
    df_sa["Products and product groups"] = prod

    lista_sa.append(df_sa)

teste_df_cpi_sa = pd.concat(lista_sa, ignore_index=True)

print("Shape CPI SA:", teste_df_cpi_sa.shape)
teste_df_cpi_sa
```

```python
def teste_cpi_trim_mes(df_mes, lim_inf=0.20, lim_sup=0.80):
    df = (
        df_mes
        .dropna(subset=["mom", "peso"])
        .sort_values("mom")
        .copy()
    )

    # pesos acumulados
    df["peso_acum"] = df["peso"].cumsum()

    # peso efetivo usado (após trimming)
    df["peso_efetivo"] = 0.0

    for i, row in df.iterrows():
        w_prev = row["peso_acum"] - row["peso"]
        w_curr = row["peso_acum"]

        # totalmente fora (abaixo)
        if w_curr <= lim_inf:
            continue

        # totalmente fora (acima)
        if w_prev >= lim_sup:
            continue

        # totalmente dentro
        if (w_prev >= lim_inf) and (w_curr <= lim_sup):
            df.at[i, "peso_efetivo"] = row["peso"]
            continue

        # cruza o limite inferior (Note 4)
        if (w_prev < lim_inf) and (w_curr > lim_inf):
            df.at[i, "peso_efetivo"] = w_curr - lim_inf
            continue

        # cruza o limite superior (Note 5)
        if (w_prev < lim_sup) and (w_curr > lim_sup):
            df.at[i, "peso_efetivo"] = lim_sup - w_prev
            continue

    # média ponderada dos 60% centrais
    trim_mom = (
        (df["mom"] * df["peso_efetivo"]).sum()
        / df["peso_efetivo"].sum()
    )

    return trim_mom
```

```python
# Série CPI-Trim (MoM)

serie_cpi_trim = (
    teste_df_base
    .groupby("REF_DATE")
    .apply(teste_cpi_trim_mes)
    .rename("cpi_trim_mom")
    .reset_index()
    .sort_values("REF_DATE")
)

serie_cpi_trim
```

```python
# Sanity Check
mes_teste = serie_cpi_trim["REF_DATE"].iloc[-1]

df_mes_teste = teste_df_base.query("REF_DATE == @mes_teste").copy()
_ = teste_cpi_trim_mes(df_mes_teste)  # roda a função para popular lógica

# reexecutar internamente para inspecionar
df = (
    df_mes_teste
    .dropna(subset=["mom", "peso"])
    .sort_values("mom")
    .copy()
)
df["peso_acum"] = df["peso"].cumsum()
df["peso_efetivo"] = 0.0

for i, row in df.iterrows():
    w_prev = row["peso_acum"] - row["peso"]
    w_curr = row["peso_acum"]
    if w_curr <= 0.20 or w_prev >= 0.80:
        continue
    if w_prev >= 0.20 and w_curr <= 0.80:
        df.at[i, "peso_efetivo"] = row["peso"]
    elif w_prev < 0.20 and w_curr > 0.20:
        df.at[i, "peso_efetivo"] = w_curr - 0.20
    elif w_prev < 0.80 and w_curr > 0.80:
        df.at[i, "peso_efetivo"] = 0.80 - w_prev

print("Peso efetivo:",  df["peso_efetivo"].sum())
```

```python
def cpi_trim_mes_detalhe(df_mes: pd.DataFrame, lim_inf=0.20, lim_sup=0.80) -> pd.DataFrame:
    df = (
        df_mes
        .dropna(subset=["mom", "peso"])
        .sort_values("mom")
        .copy()
    )

    df["peso_acum"] = df["peso"].cumsum()
    df["peso_efetivo"] = 0.0

    for i, row in df.iterrows():
        w_prev = row["peso_acum"] - row["peso"]
        w_curr = row["peso_acum"]

        # totalmente fora (abaixo)
        if w_curr <= lim_inf:
            continue

        # totalmente fora (acima)
        if w_prev >= lim_sup:
            continue

        # totalmente dentro
        if (w_prev >= lim_inf) and (w_curr <= lim_sup):
            df.at[i, "peso_efetivo"] = row["peso"]
            continue

        # cruza limite inferior (Note 4)
        if (w_prev < lim_inf) and (w_curr > lim_inf):
            df.at[i, "peso_efetivo"] = w_curr - lim_inf
            continue

        # cruza limite superior (Note 5)
        if (w_prev < lim_sup) and (w_curr > lim_sup):
            df.at[i, "peso_efetivo"] = lim_sup - w_prev
            continue

    df["included"] = df["peso_efetivo"] > 0
    soma_eff = df.loc[df["included"], "peso_efetivo"].sum()

    # share dentro do "miolo" (60% do basket)
    df["share"] = np.where(df["included"], df["peso_efetivo"] / soma_eff, 0.0)

    # MoM SA em %
    df["mom_pct"] = df["mom"] * 100

    # contribuição em bps para o CPI-trim do mês
    # (mom decimal * share) * 10000
    df["contrib_bps"] = (df["mom"] * df["share"]) * 10000

    return df

mes_ref = pd.to_datetime(serie_cpi_trim["REF_DATE"].max())

df_mes = teste_df_base.query("REF_DATE == @mes_ref").copy()

det = cpi_trim_mes_detalhe(df_mes, lim_inf=0.20, lim_sup=0.80)

# Só incluídos
det_inc = det.loc[det["included"]].copy()

print("Mês:", mes_ref.date())
print("Incluídos (n):", det_inc.shape[0])
print("Soma peso_efetivo:", det_inc["peso_efetivo"].sum().round(6))

# CPI-trim m/m (sanity)
cpi_trim_mom_mes = (det_inc["mom"] * det_inc["share"]).sum()
print("CPI-Trim m/m do mês (decimal):", cpi_trim_mom_mes)
print("CPI-Trim m/m do mês (%):", 100*cpi_trim_mom_mes)

```

```python
# TABELA 1 — Single-month components included (MoM SA)

tabela_mom = (
    det_inc[["Products and product groups", "mom_pct"]]
    .sort_values("mom_pct", ascending=False)
    .rename(columns={"Products and product groups": "component"})
    .reset_index(drop=True)
)

# TABELA 2 — contribuição em bps (ordena por contrib_bps)
tabela_contrib = (
    det_inc[["Products and product groups", "contrib_bps", "share", "mom_pct"]]
    .sort_values("contrib_bps", ascending=False)
    .rename(columns={"Products and product groups": "component"})
    .reset_index(drop=True)
)

display(tabela_mom.head(30))
display(tabela_contrib.head(30))


TOP_N = 30

# seleção para MoM: top positivos e top negativos (para ficar “equilibrado”)
mom_sorted = tabela_mom.copy()
mom_pos = mom_sorted[mom_sorted["mom_pct"] >= 0].head(TOP_N//2)
mom_neg = mom_sorted[mom_sorted["mom_pct"] < 0].tail(TOP_N//2)
mom_plot = pd.concat([mom_pos, mom_neg], ignore_index=True).sort_values("mom_pct")

#seleção para contrib: top contribuições absolutas
contrib_plot = tabela_contrib.copy()
contrib_plot["abs"] = contrib_plot["contrib_bps"].abs()
contrib_plot = contrib_plot.sort_values("abs", ascending=False).head(TOP_N).sort_values("contrib_bps")

plt.figure(figsize=(16, 7))

# painel esquerdo
ax1 = plt.subplot(1, 2, 1)
ax1.barh(mom_plot["component"], mom_plot["mom_pct"])
ax1.axvline(0, linewidth=1)
ax1.set_title("Single-month components included in CPI-Trim (MoM SA)")
ax1.set_xlabel("Percent (m/m)")
ax1.set_ylabel("")
ax1.grid(axis="x", alpha=0.25)

# painel direito
ax2 = plt.subplot(1, 2, 2)
ax2.barh(contrib_plot["component"], contrib_plot["contrib_bps"])
ax2.axvline(0, linewidth=1)
ax2.set_title("Single-month components contribution to CPI-Trim (bps)")
ax2.set_xlabel("bps")
ax2.set_ylabel("")
ax2.grid(axis="x", alpha=0.25)

plt.suptitle(f"Canadá — CPI Trimmed Mean ({mes_ref:%b-%Y})", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```
